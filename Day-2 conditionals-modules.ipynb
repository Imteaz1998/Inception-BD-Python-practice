# IF-ELSE Conditional

```python
"""
Actual credentials

Email: xyz@gmail.com
Pass: 123456
"""

email=input("Enter your mail: ")
password=input("Enter your password")

actual_email="xyz@gmail.com"
actual_pass="Imteaz@123456"

if email==actual_email and password==actual_pass:
    print("Login successfull")
else:
    print("Invalid credentials")
```

# IF-ELIF-ELSE conditional

```python
"""
Actual credentials

Email: xyz@gmail.com
Pass: 123456
"""

email=input("Enter your mail: ")
password=input("Enter your password")

actual_email="xyz@gmail.com"
actual_pass="Imteaz@123456"

if email==actual_email and password==actual_pass:
    print("Login successfull")
elif email == actual_email or password != actual_pass:
    print(" password wrong please check")
else:
    print("Invalid credentials")
```

# Nested conditional statments

```python
"""
Actual credentials

"""

email=input("Enter your mail: ")
password=input("Enter your password")

actual_email="xyz@gmail.com"
actual_pass="Imteaz@123456"

if email==actual_email and password==actual_pass:
    print("Login successfull")
elif email == actual_email or password != actual_pass:
    print("Please try again")
    re_pass=input("Input password again")
    if re_pass==actual_pass:
        print("Success")
    else:
        print("Hobe na Hobe na")
else:
    print("Invalid credentials")
```
