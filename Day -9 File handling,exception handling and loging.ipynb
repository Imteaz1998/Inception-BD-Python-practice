Multiline

```python
f=open("files/sample2.txt","w")
f.write("Hello Python")
f.write("\nHello Python")
f.write("\nHello Python")
f.close()
```

```python
# Ovewriting previous content
f=open("files/sample2.txt","w")
f.write("Bye Python")
f.write("\nBye Python")
f.write("\nBye Python")
f.close()
```

```python
# append mode

f=open("files/sample2.txt","a")
f.write("Hello Python")
f.write("\nHello Python")
f.write("\nHello Python")
f.close()
```

```python
# write lines
L=["hell\n","hi\n",'how are you\n']
f=open("files/sample2.txt","w")
f.writelines(L)
f.close()
```

# File Handling using context Manager(With)
* Automatic close

```python
with open("files/sample2.txt","w") as f:
    f.write("Hello this is a paragraph")
```

```python
with open("files/sample2.txt","r") as f:
    s=f.read()
    print(s)
```

```python
# cursor setting

with open("files/sample2.txt","r") as f:
    f.seek(5)
    print(f.read())
```

```python
# Tells where is the cursor
with open("files/sample2.txt","r") as f:
    print(f.tell())
```

# logging in python

```python
import logging
import os

LOG_DIR="logs"
LOG_FILE_NAME="application.log"

os.makedirs(LOG_DIR,exist_ok=True) # True means jodi thake o ignore koro

log_path=os.path.join(LOG_DIR,LOG_FILE_NAME)
```

```python
print(log_path)
```

```python
#writing logging config
logging.basicConfig(
    filename=log_path,
    format="[ %(asctime)s ] %(name)s - %(levelname)s - %(message)s",
    level=logging.INFO
)
```

```python
def profile(name):
    if type(name) == str:
        print(f"Welcome {name}")
    else:
        logging.info("Your input is not valid")
```

```python
profile("Imteaz")
```
