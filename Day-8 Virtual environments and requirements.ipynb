command for venv
(base)-> Default environement
pip list -> to check packages
conda create -n helmet python=3.9 -y --> creating environment
conda activate helmet --> to activate environment
conda deactivate --> to deactivate environment
conda remove --name helmet --all --> to remove the env file (fiirst deactivate)
conda env list --> to show all the environment
conda create --prefix .\helmet python=3.9 -y -- > To create environment in particular directory go to that directory and run command
conda activate .\helmet --> to activate that env
pip install -r requirements.txt-> to activate packages
