# Type Conversion in Python

1. Implicit- Done by python
2. Explicit- Done by the programmer

```python
#Implicit
a=5+5.5
print(a)
print(type(a))
```

```python
#Explicit

b=4+int("4")
print(b)
print(type(b))
```

# Operators in Python
1. Arithmatic operators
2. Relational operators
3. Logical operators
4. Bitwise operators
5. Assignment operators
6. Membership operators

```python
#Arithmatic operators +.-,*,/,%

a=5
b=2
c=a%b
print(c)
```

```python
# Relational Operators >,<,>=,<=,==,!=

print(4>5)
```

```python
# Logical Operators and,or,not
print(0 and 1)
```

```python
print(0 or 1)
```

```python
#Bitwise operators &, | , ^(XOR), ~(Bitwise not), right shift(>>), left shift(<<)

print(2&3)
print(2|3)
print(2^3)
print(~3)
print(4>>2)
print(4<<2)
```

```python
#Assignment operators
a=2
a+=1
print(a)
```

```python
#Membership Operators in / not in

print('l' in 'Bangladesh')
print('w' in 'Bangladesh')

```
