# *args and *kwargs

```python
# Unlimited number of arguments

def multiply(*args):
    product=1
    for i in args:
        product=product*i
    return product
mul=multiply(1,2,3)
mul
```

* map()
The map() function in Python applies a given function to each item in an iterable (like a list, tuple, or set) and returns an iterator that yields the results. It's a powerful tool for performing element-wise transformations on data. 
Here's a breakdown of its functionality:
Syntax: map(function, iterable, ...)
function: The function to be applied to each item. This function can be a built-in function, a user-defined function, or a lambda function.
iterable: One or more iterables whose elements will be processed by the function. If multiple iterables are provided, the function must accept a corresponding number of arguments.
Operation: map() iterates through the provided iterable(s), applying the function to each element (or corresponding elements from multiple iterables). It does not modify the original iterable(s).
Return Value: map() returns a map object, which is an iterator. This object can then be converted into a list, tuple, or other collection type if needed (e.g., list(map_object)).

```python
L=[1,2,3,4]
list(map(lambda x:x**4, L))
```

* filter
The built-in filter() function in Python is used to construct an iterator from elements of an iterable for which a function returns true. In essence, it filters out elements that do not satisfy a specific condition.
Here's how it works:
Arguments: It takes two arguments:
A function: This function is applied to each item of the iterable. It should return a boolean value (True or False).
An iterable: This is the sequence (like a list, tuple, or string) whose elements you want to filter.
Filtering Logic: For each element in the iterable, the function is called with that element as an argument.
If the function returns True, the element is included in the filtered result.
If the function returns False, the element is excluded.
Return Value: filter() returns a filter object, which is an iterator. This object can then be converted into a list, tuple, or other iterable type if needed.

```python
L=[1,2,3,4]
list(filter(lambda x:x>2, L))
```

* reduce
The reduce() function in Python, found within the functools module, cumulatively applies a given function to the items of an iterable (like a list, tuple, or set), thereby reducing the iterable to a single, cumulative value.
Here's how it works:
Takes a function and an iterable: reduce() requires a function that accepts two arguments and an iterable as input.
Cumulative application: It applies the provided function to the first two items of the iterable, generating a result.
Iterative reduction: This result then becomes the first argument for the next application of the function, with the third item of the iterable serving as the second argument. This process continues, iteratively combining the accumulated result with the next item in the iterable, until only a single value remains.
Optional initializer: An optional third argument, initializer, can be provided. If present, it acts as the initial value for the reduction, preceding the items in the iterable.

```python
import functools
L=[1,2,3,4,5]

functools.reduce(lambda x,y:x+y,L)
```

* Enumerate
The enumerate() function in Python is a built-in function that adds a counter to an iterable (like a list, tuple, or string) and returns an enumerate object. This object yields pairs of (index, element) for each item in the iterable, making it convenient to iterate through a sequence while simultaneously accessing both the element and its corresponding index. 
Key functionalities of enumerate():
Provides a counter: It automatically generates a counter (starting from 0 by default, or a specified start value) for each item in the iterable.
Simplifies iteration with indexes: It eliminates the need for manually managing an index variable within a loop, leading to more concise and readable code.
Returns (index, element) pairs: When iterated, it yields tuples where the first element is the index and the second element is the value from the original iterable.

```python
L=["Imteaz","Ahmed","Minhaz"]

for i,n in enumerate(L,start=1):
    print(i,n)
```
