# Difference between tuples and lists
* Syntax
* Mutability
* Speed
* Memory
* Built in functionality
* Error
* Usability

```python
# Memory efficiency
import sys
L=list(range(1000))
T=tuple(range(1000))

print('List size', sys.getsizeof(L))

print('Tuple size', sys.getsizeof(T))
```

```python
# Mutability

a=[1,2,3,4]
b=a
a.append(4)
print(a)
print(b)
print(id(a))
print(id(b))
```

```python
a=(1,2,3)
b=a
a=a+(4,)
print(a)
print(b)
```

# Dictionary comprehension

```python
{i:i**2 for i in range(1,11)}
```

```python
products={'phone':10 ,'laptop': 0, 'charger': 32}
for i,j in products.items():
    if j>0:
        print(i,j)
```

```python
{key:value for (key,value) in products.items() if value>0}
```

```python
# Using ZIP
days=["Sunday","Monday","Tuesday","Wednesday","Thursday"]
temp_c=[33.5,32.6,29.0,38.0,40.00]
{i:j for (i,j) in zip(days,temp_c)}
```
