Project Guessing game
1 Generate a random number
2 Take an input from user
3 compare/match the number
4 If guess> generate number-> guess a lower number
5 if guess< generated number- guess a higher number
6 if number correct show the number

```python
import random

jackpot=random.randint(1,100)
guess_number=int(input("Enter a number: "))
counter=1

while guess_number!=jackpot:

    if guess_number>jackpot:
        print("Try to guess lower")
    else:
        print("Wrong!, Try to guess higher")

    guess_number=int(input("Enter a number: "))
    counter+=1
else:
    print("Voila! correct guess")
    print("Your attempt is:", counter)

    


    


```

For Loop

```python
for i in range(1,11):
    print(i)
```

```python
for i in range(10,0,-1): # here -1 is range and reverse
    print(i)
```

```python
for i in range(1,11,2): #Skip value 2
    print(i)
```

```python
for i in  "Imteaz":
    print(i)
```

Nested Loops

```python
for i in range(1,5):
    for j in range(1,5):
        print(i,j)
```

```python
# print traingle coordinates

row=int(input("Input the number of rows: "))
for i in range(1,row+1):
    for j in range(1,i+1):
        print("*",end="")
    print("")
```

# Loop Control Statements
1. Break
2. Continue
3. Pass

```python
for i in range(1,11):
    if i==6:
        break
    print(i)
```

```python
# continue
for i in range(1,10):
    if i ==5:
        continue
    print(i)
```

```python
#Pass
for i in range(1,11):
    pass
```
