# List comprehension
* newlist= [expression for item in iterable if condition == True]

```python
# ADD one number to list

l=[]

for i in range(1,11):
    l.append(i)
print(l)
```

```python
# list comprehension form
lc=[5 for i in range(1,11)]
```

```python
lc
```

```python
# Scaler multiplication in a vector

v=[1,3,4]
s=-3
l=[]

for i in v:
    result=i*s
    l.append(result)
print(l)
```

```python
Lc=[i*s for i in v]
print(Lc)
```

```python
# print all numbers if divisable by 5

for i in range(1,51):
    if i%5==0:
        print(i)
```

```python
lc=[i for i in range (1,51) if i%5==0]
```
