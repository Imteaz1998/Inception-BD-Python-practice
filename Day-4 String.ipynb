# Write a program to identify palindrome

```python
word=input("enter a word: ")
flag=True

for i in range(0, len(word)//2):
    if word[i] != word[len(word)-i-1]:
        flag=False
        print("Not a palindorme")
        break
if flag:
    print("Palindrome")
```
